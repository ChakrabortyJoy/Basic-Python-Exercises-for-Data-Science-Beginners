```python

#Subplot()
# Compute the x and y coordinates for points on sine and cosine curves 
x = np.arange(0, 5 * np.pi, 0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  
   
# Set up a subplot grid that has height 2 and width 1, 
# and set the first such subplot as active. 
plt.subplot(2, 1, 1)
   
# Make the first plot 
plt.plot(x, y_sin,'r--') 
plt.title('Sine')  
   
# Set the second subplot as active, and make the second plot. 
plt.subplot(2, 1, 2) 
plt.plot(x, y_cos,'g--') 
plt.title('Cosine')  
   
# Show the figure. 
plt.show()
```

```python
## bar plot

x=[2,8,10]
y=[11,16,9]

x1=[3,9,11]
y1=[6,15,7]
plt.bar(x,y,color='b')
plt.bar(x1,y1,color='r')
plt.title("Bar graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
```
